#### Preprocess the housing dataset by encoding all categorical variables using appropriate techniques. Ensure the final dataset is fully numeric and ready for machine learning analysis. y = SalePrice.

```python
import pandas as pd
import warnings 
warnings.filterwarnings('ignore') 
```

```python
dataframe = pd.read_csv('house.csv')
dataframe.head()
```

```python
df = dataframe.copy()
df
```

```python
df.info()
```

```python
df = df.drop(['Alley','MasVnrType','FireplaceQu','PoolQC','Fence','MiscFeature'], axis=1)
df
#I have droped this columns because of too much null values
```

```python
df = df.drop('Utilities',axis=1)
df.head()
#I have droped this because all of them are same value
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
categorical_columns = df.select_dtypes('object').columns
categorical_columns
```

```python
for c in categorical_columns:
    if c == 'SaleType' or c == 'Exterior1st' or c=='Exterior2nd' or c== 'Neighborhood':
        continue #I will use binary encoding for these columns
    elif c=='LotConfig':
        continue #I will use one hot encoding for this column
    df[c] = le.fit_transform(df[[c]])

df.head()
```

```python
import category_encoders as ce
BinEn = ce.BinaryEncoder(cols=['SaleType','Exterior1st', 'Exterior2nd', 'Neighborhood'])
BinEn
```

```python
df = BinEn.fit_transform(df)
df.head()
```

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
lb = LabelBinarizer()
area_encoded = lb.fit_transform(df['LotConfig'])

area_encoded_df = pd.DataFrame(area_encoded, columns=lb.classes_)
area_encoded_df.head()
```

```python
area_encoded_df = area_encoded_df.drop(columns=lb.classes_[0])
area_encoded_df.head()
```

```python
df = pd.concat([df.drop('LotConfig', axis=1), area_encoded_df], axis=1)
df.head()
```

## Filling null values with median

```python
columns_with_nulls = df.columns[df.isna().any()]
columns_with_nulls
```

```python
for col in columns_with_nulls:
    med = df[col].median()
    df[col] = df[col].fillna(med)

df.head()
```

```python
df2 = df['SalePrice']
df2.head()
```

```python
df = df.drop("SalePrice",axis=1)
df = pd.concat([df,df2],axis=1)
df.head()
```

```python
df.info()
```

```python
X = df.iloc[:, :-1]
X
```

```python
Y = df['SalePrice']
Y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,Y)
```

```python
reg.score(X,Y)
#For all level encoder This result was 0.84
```
