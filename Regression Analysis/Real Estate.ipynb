### Dataset Explanation

1. **X1 transaction date** (`int32`):  
   - Represents the year of the transaction, indicating when the real estate sale occurred.  
   - Example: 2012 or 2013.

2. **X2 house age** (`float64`):  
   - The age of the house in years at the time of the transaction.  
   - Example: 32.0 means the house is 32 years old.

3. **X3 distance to the nearest MRT station** (`float64`):  
   - The distance from the house to the nearest MRT station, measured in meters.  
   - Example: 84.87882 meters.

4. **X4 number of convenience stores** (`int64`):  
   - The number of convenience stores within a certain vicinity of the house.  
   - Example: 10 means there are 10 convenience stores nearby.

5. **Y house price of unit area** (`float64`):  
   - The price per unit area (e.g., per square meter) of the house during the transaction.  
   - Example: 37.9 means 37.9 currency units per square meter.

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import  mean_squared_error
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import PolynomialFeatures
import warnings
warnings.filterwarnings('ignore')
```

#### **1. Split the Dataset**
- Split the data into **training (70%)** and **testing (30%)** subsets.
- Use the `train_test_split` function to ensure randomness in the split.
- Define:
  - **Features (X)**: `X1 transaction date`, `X2 house age`, `X3 distance to the nearest MRT station`, `X4 number of convenience stores`.
  - **Target (Y)**: `Y house price of unit area`.

```python
df.head()
```

```python
X = df[["X1 transaction date","X2 house age","X3 distance to the nearest MRT station","X4 number of convenience stores" ]]
Y = df["Y house price of unit area"]
```

```python
xtrain, xtest, ytrain, ytest = tts(X, Y, train_size=.70, random_state=3)
xtrain
```

#### **2. Check Correlation**
- Compute the correlation matrix to understand the relationships between features and the target variable.
- Visualize the correlation matrix using a **heatmap**:
  - Strong correlations between features and the target suggest potential predictors.
  - Identify multicollinearity (strong correlations between features), which might affect regression models.

```python
df.corr()
```

```python
import seaborn as sns

corr = df.corr()

plt.figure(figsize=(10, 8))

sns.heatmap(
    corr, 
    annot=True,             
    fmt=".2f",              
    cmap="coolwarm",        
    cbar=True,            
    square=True,            
    linewidths=0.5,         
)

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.title("Correlation Heatmap")   
plt.tight_layout()                     
plt.show()
```

X3 (distance to the nearest MRT station) and X4 (number of convenience stores): medium negative correlation (-0.603).
Other pairwise correlations are low. So, there is no significant multicollinearity among the features.

#### **3. Apply Regression Models**

**a. Linear Regression**
1. Train a **Linear Regression** model using the training data.
2. Predict the target values using the testing data.

**b. Polynomial Regression**
1. Use `PolynomialFeatures` to generate polynomial terms for the features (e.g., degree = 2 or 3).
2. Train a **Linear Regression** model on the transformed polynomial features.
3. Predict the target values using the transformed testing data.

```python
reg = LinearRegression()
reg.fit(xtrain,ytrain)
```

```python
y_pred = reg.predict(xtest)

dftest = xtest.copy()
dftest['Y house price of unit area'] = ytest 
dftest['Predicted Y for degree 1'] = y_pred

dftest.head()
```

```python
reg.score(xtest, ytest)
```

```python
reg.score(xtrain, ytrain)
```

```python
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(xtrain)
x_poly_test = poly.fit_transform(xtest)
x_poly.shape
x_poly_test.shape
```

```python
reg_poly = LinearRegression()
reg_poly.fit(x_poly, ytrain)
```

```python
dftest['Predicted Y by degree 2'] = reg_poly.predict(poly.transform(xtest))
dftest.head()
```

```python
reg_poly.score(x_poly_test, ytest)
```

```python
reg_poly.score(x_poly,ytrain)
```

#### **4. Evaluate Model Performance**
- Evaluate both Linear and Polynomial Regression models using:
  1. **Mean Squared Error (MSE)**:
     - Measures the average squared difference between predicted and actual values.
     - Lower MSE indicates better performance.
  2. **R-squared (R²)**:
     - Measures how well the model explains the variability of the target variable.
     - Higher R² indicates better performance.
- Compare the results to determine which model performs better.

```python
mse_multi = mean_squared_error(dftest['Predicted Y for degree 1'], dftest['Y house price of unit area'])
mse_multi
```

```python
mse_poly = mean_squared_error(dftest['Predicted Y by degree 2'], dftest['Y house price of unit area'])
mse_poly
```

According to MSE the polynomial model performs better

```python
reg.score(xtest, ytest)
```

```python
reg.score(xtrain, ytrain)
```

```python
reg_poly.score(x_poly_test, ytest)
```

```python
reg_poly.score(x_poly, ytrain)
```

According to R-squared value the polynomial model performs better than the linear model

#### **5. Visualize Results**
- Plot **actual vs. predicted values** for both Linear and Polynomial Regression.
- For Polynomial Regression, visualize the curve fitting to see how well it models the data.

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.scatter(ytest, reg.predict(xtest), color='blue', alpha=0.7, label='Predicted')
plt.plot(ytest, ytest, color='red', linewidth=2, label='True Values')
plt.title('Linear Regression Predictions', fontsize=14)
plt.xlabel('Actual Target Values', fontsize=12)
plt.ylabel('Predicted Values', fontsize=12)
plt.grid(alpha=0.4)
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(ytest, reg_poly.predict(poly.transform(xtest)), color='green', alpha=0.7, label='Predicted')
plt.plot(ytest, ytest, color='red', linewidth=2, label='True Values')
plt.title('Polynomial Regression Predictions', fontsize=14)
plt.xlabel('Actual Target Values', fontsize=12)
plt.ylabel('Predicted Values', fontsize=12)
plt.grid(alpha=0.4)
plt.legend()

plt.tight_layout()
plt.show()

```

```python
X_1 = dftest[['X1 transaction date']].values  
X_2 = dftest[["X2 house age"]].values  
X_3 = dftest[['X3 distance to the nearest MRT station']].values  
X_4 = dftest[['X4 number of convenience stores']].values  
y = dftest["Predicted Y by degree 2"].values
df_sorted = dftest.sort_values('X2 house age')


poly = PolynomialFeatures(degree=2)  
X_poly_1 = poly.fit_transform(X_1)
X_poly_2 = poly.fit_transform(X_2)
X_poly_3 = poly.fit_transform(X_3)
X_poly_4 = poly.fit_transform(X_4)

curve_1 = LinearRegression()
curve_2 = LinearRegression()
curve_3 = LinearRegression()
curve_4 = LinearRegression()
curve_1.fit(X_poly_1, y)
curve_2.fit(X_poly_2, y)
curve_3.fit(X_poly_3, y)
curve_4.fit(X_poly_4, y)

t_1 = np.arange(2012, 2013.1, 0.1).reshape(-1, 1)
t_2 = np.arange(0, 43, 1).reshape(-1, 1)
t_3 = np.arange(0, 6500, 10).reshape(-1, 1)
t_4 = np.arange(0, 10.1, 0.1).reshape(-1, 1)

t_1_p = curve_1.predict(poly.transform(t_1))
t_2_p = curve_2.predict(poly.transform(t_2))
t_3_p = curve_3.predict(poly.transform(t_3))
t_4_p = curve_4.predict(poly.transform(t_4))


```
